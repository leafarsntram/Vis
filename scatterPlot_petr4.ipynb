```python
import plotly.express as px
import pandas as pd

df = pd.read_html('http://github.com/leafarsntram/Vis/blob/master/PETR4c.csv')[0]

fig = px.box(df, x="Date", y="volume", points="all")
fig.update_layout(
    #barmode='stack',
    title="Ação PETR4 em 2019 ",
    xaxis_title="Ações",
    margin=dict(l=20, r=20, t=50, b=10),
    #yaxis_title="Preço",
    font=dict(
        family="Arial, monospace",
        size=18,
        color="#7f7f7f"
    )
)

fig.show()
```
