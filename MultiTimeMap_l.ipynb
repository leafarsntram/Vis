```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
import json
from urllib.request import urlopen
```

```python
### carregando dados web###
##lendo e decodificando geojson##
with urlopen('https://raw.githubusercontent.com/datalivre/Conjunto-de-Dados/master/br_states.json') as response:
    states = json.load(response)
    
##Carregando dados csv##
df = pd.read_csv("https://raw.githubusercontent.com/leafarsntram/Vis/master/data/series_vaca.csv", delimiter=";"
                , dtype={"UF": str})
df.rename(columns={'UF': 'id'}, inplace=True)
```

```python
# create figure
fig = go.Figure()

years = ['2014','2015','2016']    
data = [] 
fig.update_layout(title_text="Vacas Ordenhadas no Brasil por UF (fonte:IBGE)")
# Add surface trace
fig.add_trace(go.Choroplethmapbox(geojson=states, 
                                    locations=df.id, 
                                    z=df['2016'],
                                    colorscale="Viridis", 
                                    zmin=max(df['2016']), 
                                    zmax=min(df['2016']),
                                    marker_opacity=0.5, 
                                    marker_line_width=0 
                                 )
             )
fig.add_trace(go.Choroplethmapbox(geojson=states, 
                                    locations=df.id, 
                                    z=df['2015'],
                                    colorscale="Viridis", 
                                    zmin=max(df['2015']), 
                                    zmax=min(df['2015']),
                                    marker_opacity=0.5, 
                                    marker_line_width=0
                                 )
             )
fig.add_trace(go.Choroplethmapbox(geojson=states, 
                                    locations=df.id, 
                                    z=df['2014'],
                                    colorscale="Viridis", 
                                    zmin=max(df['2014']), 
                                    zmax=min(df['2014']),
                                    marker_opacity=0.5, 
                                    marker_line_width=0
                                 )
             )
#define layout    
fig.update_layout( 
                    width=800,
                    height=900,
                    autosize=False,
                    margin={"r":0,"t":0,"l":0,"b":0},
                    mapbox_center = {"lat": -16, "lon": -50},
                    mapbox_style="carto-positron",
                    mapbox_zoom=3,
                    title = "Vacas Ordenhadas no Brasil por UF"
                    )



fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list([
                dict(label="2016",
                     method="update",
                     args=[{"visible": [True, False, False]},
                           {"title": "2016" #,"annotations": []
                           }]),
                dict(label="2015",
                     method="update",
                     args=[{"visible": [False, True, False]},
                           {"title": "2015" #, "annotations": high_annotations
                           }]),
                dict(label="2014",
                     method="update",
                     args=[{"visible": [False, False, True]},
                           {"title": "2014"#, "annotations": low_annotations
                           }]),
                
            ]),
        )
    ])

# Add annotation
fig.update_layout(
                annotations=[
                    dict(text="Trace type:", showarrow=False,
                        x=0, y=1.085, yref="paper", align="left")
                ]
            )

fig.show()
```
