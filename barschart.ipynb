```python
import plotly.graph_objects as go

x = ['1950','1960','1970','1980','1991','2000','2010']
y = [51944397,70992343,94508583,121150573,146917459,169590693,190755799]

# Use the hovertext kw argument for hover text
fig = go.Figure(data=[go.Bar(x=x, y=y)])# , hovertext=['27% market share', '24% market share', '19% market share'])])
# Customize aspect
fig.update_traces(marker_color='rgb(158,202,225)', marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)
fig.update_layout(title_text='Crescimento populacional Brasileiro')
fig.show()
```
