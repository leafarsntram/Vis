```python
import plotly.graph_objects as go
import pandas as pd
from datetime import date, datetime

# Add data
df = pd.read_html('http://github.com/leafarsntram/Vis/blob/master/data/ITUB4.SA.csv')[0]
df2 = pd.read_html('http://github.com/leafarsntram/Vis/blob/master/data/JBSS3.SA.csv')[0]
df3 = pd.read_html('http://github.com/leafarsntram/Vis/blob/master/data/ABEV3.SA.csv')[0]

fig = go.Figure()
# Create and style traces
fig.add_trace(go.Scatter(x=df.Date, y=df.Close, name='ITUB4',
                         line=dict(color='firebrick', width=4)))
fig.add_trace(go.Scatter(x=df2.Date , y=df2.Close, name = 'JBSS3',
                         line=dict(color='royalblue', width=4)))
fig.add_trace(go.Scatter(x=df3.Date, y=df3.Close, name='ABEV3',
                         line=dict(color='gray', width=4,
                              dash='dot') # dash options include 'dash', 'dot', and 'dashdot'
))
#fig.add_trace(go.Scatter(x=df.Date, y=df.Close, name='ITUB4',                         line = dict(color='royalblue', width=4, dash='dash')))
#fig.add_trace(go.Scatter(x=df2.Date, y=df2.Close, name='JBSS3',                         line = dict(color='firebrick', width=4, dash='dot')))
#fig.add_trace(go.Scatter(x=df3.Date, y=df3.Close, name='ABEV3',                         line=dict(color='royalblue', width=4, dash='dot')))

# Edit the layout
fig.update_layout(title='Preo Aes 2019',
                   xaxis_title='MS',
                   yaxis_title='Preo R$')


fig.show()
```
